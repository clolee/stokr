# matplotlib 가로선 세로선

```python
%matplotlib inline
import matplotlib.pylab as plt

plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['axes.grid'] = True 
```

```python
import pandas as pd
from datetime import datetime
from pandas_datareader import data

start = datetime(2016, 1, 1)
end = datetime(2016, 3, 31)

df = data.get_data_yahoo('^KS11', start, end)
```

# 가로선, 세로선
특정 시점 혹은 가격을 표현하고자 할 때, 가로선 혹은 세로선을 표현하고 싶을때

* http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.axvline
* http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.axvline    

```python
# 특정 가격(2000 포인트)을 설명하기 위한 가로선 

plt.plot(df['Adj Close'])
plt.axhline(y=2000, color='r', linewidth=2)
plt.show()
```

```python
# 트정 시점을 표시하기 위한 세로선

plt.plot(df['Adj Close'])
plt.axvline(x=datetime(2016, 2,1), color='r', linestyle='--', linewidth=3)
plt.show()
```
