## plot.ly 데이터 시각화

```python
%matplotlib inline
from IPython.display import display
```

```python
import pandas as pd
import numpy as np
```

```python
df_master = pd.read_csv('data/marcap-all.csv')
df_master.set_index('code', inplace=True, drop=False)
df_master.head()
```

```python
# 관심 종목 (favorite stocks)
fav_code_corp = [
    ('005930', '삼성전자'),
    ('005380', '현대차'),
    ('015760', '한국전력'),
    ('028260', '삼성물산'),
    ('012330', '현대모비스'),
    ('090430', '아모레'),
    ('000660', '하이닉스'),
    ('051910', 'LG화학'),
    ('032830', '삼성생명'),
    ('000270', '기아차'),
    ('035420', 'NAVER'),
    ('018260', '삼성SDS'),
    ('055550', '신한지주'),
    ('017670', 'SK텔레콤'),
    ('034730', 'SK'),
    ('051900', 'LG생활건강'),
    ('005490', 'POSCO'),
    ('033780', 'KT&G'),
    ('000810', '삼성화재'),
    ('105560', 'KB금융'),
]

fav_stocks = pd.DataFrame(data=fav_code_corp , columns=['code', 'corp_name'])
fav_stocks

# 이름을 관심종목(fav_stocks)에 동일하게 맞춘다.
for index, row in fav_stocks.iterrows():
    df_master.loc[row['code'], 'corp_name'] = row['corp_name']

# 선두 7개 종목의 코드만 추출
top7_codes = fav_stocks[:7]['code']
top7_codes

# 관심종목에 해당하는 종목만 추출
df_fav = df_master.loc[df_master['code'].isin(top7_codes)]

# 피봇
pivoted = df_fav.pivot('year', 'corp_name', 'rank')
pivoted

#fav_stocks[:7]['corp_name'].values

##  상위 7개 관심종목(fav_stocks) 의 종목명
cols = fav_stocks[:7]['corp_name']
print (cols)

# 컬럼 순서를 관심종목(fav_stocks) 순서로 바꾸기
pivoted = pivoted[cols]
```

```python
## plotly chart

import plotly.plotly as py
import plotly.graph_objs as go

data = []
for col in pivoted.columns:
    trace =  go.Scatter(
        x = pivoted.index,
        y = pivoted[col],
        name = col,
    )
    data.append(trace)

layout = go.Layout(
    title='시총상위 7종목의 과거 20년간 순위 변화)',
    yaxis=dict(
        autorange='reversed',
        title='순위',
    ),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marcap/spike01')
```

```python
# 시총상위 20개 종목의 코드
top20_codes = fav_stocks[:20]['code']

# 전체에서 시총상위 20개 종목(top20_codes) row들을 추출하고 인덱스 지정
df_fav = df_master.loc[df_master['code'].isin(top20_codes)]
df_fav.set_index('code', inplace=True)

# 년도(시간), 종목명, 순위로 피벗
pivoted = df_fav.pivot('year', 'corp_name', 'rank')

# 컬럼 순서를 관심종목(fav_stocks) 순서로 바꾸기
pivoted = pivoted[fav_stocks[:20]['corp_name']]
pivoted.head(20)
```

```python
## plotly chart

import plotly.plotly as py
import plotly.graph_objs as go

data = []
for col in pivoted.columns:
    trace =  go.Scatter(
        x = pivoted.index,
        y = pivoted[col],
        name = col,
    )
    data.append(trace)

layout = go.Layout(
    title='시총상위 20 종목의 과거 20년간 순위 변화',
    yaxis=dict(
        autorange='reversed',
        title='순위',
    ),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marcap/spike02')
```

## 2015년 분석

```python
df2 = df_master[df_master['year'] == '2015-12-01']

display (df2.head(10))
display (len(df2))
```

```python
# value_counts() Series내의 values 항목과 항목의 개수

sector_counts = df2['sector'].value_counts()

display (sector_counts.count())  # 업종수 
display (sector_counts.index)    # 업종항목: '화학', '서비스업', '반도체', 'IT부품', 등 …
display (sector_counts.values)  # 업종내 기업수: 160 151 109 109  76  64  64  …
```

## 업종 분포 차트

```python
top20 = sector_counts[:20]
top20
```

```python
display (top20.index)
display (top20.values)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

colors_list = [ "#C41F3B", "#FF7D0A", "#ABD473", "#69CCF0", "#00FF96", "#F58CBA", "#FFFFFF", "#FFF569", "#0070DE", "#9482C9", "#C79C6E" ]

data = [
    go.Bar(
        x=top20.values[::-1],
        y=top20.index[::-1],
        orientation = 'h',
        marker=dict(
            color=colors_list * 2,
            line=dict(
                color='black',
                width=1.0
            ),
        ),
    )
]

layout = go.Layout(
    title='2015년 업종별 종목수',
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marcap/sector_marcap00')    
```

## 특정 섹터만 추출

```python
# 2015년 시가총액
print( '시가총액: ', df2['marcap'].sum() )

# 2015년 전체(df2)중에 반도체 종목만
df_semi = df2[df2['sector']=='반도체']
display( df_semi.head() )

# 반도체 업종 종목수
print( '반도체 업종 종목 수: ', len(df_semi) ) 

# 반도체 업종 시가총액
print( '반도체 업종 시가총액: ', df_semi['marcap'].sum() ) 

# 반도체 업종 시가총액 비중
print( '반도체 업종 시가총액 비중: ', df_semi['marcap'].sum() / df2['marcap'].sum() )
```

```python
# 피벗테이블 (marcap)
piv = pd.pivot_table(df2, values='marcap', index=['sector'], aggfunc=np.sum)

# 복사본을 만들고, 값 기준으로 내림차순 소트, 상위 20개 업종 표시
sector_marcap = piv.copy()
sector_marcap.sort_values(ascending=False, inplace=True)
sector_marcap[:20]
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

data = [
    go.Bar(
        x=sector_marcap[:20].values[::-1],
        y=sector_marcap[:20].index[::-1],
        orientation = 'h',
    )
]

layout = go.Layout(
    title='2015년 업종별 시가총액 (단위:조원)',
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marcap/sector_marcap01')
```
