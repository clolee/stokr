# matplotlib 차트 채우기

```python
%matplotlib inline
import matplotlib.pylab as plt

plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['axes.grid'] = True 
```

```python
import pandas as pd
from datetime import datetime
from pandas_datareader import data

start = datetime(2016, 1, 1)
end = datetime(2016, 3, 31)

df = data.get_data_yahoo('^KS11', start, end)
```

# 채우기
* http://matplotlib.org/examples/pylab_examples/fill_between_demo.html

변화를 강조

* 최다대/최소 가격으로 y축의 최대/최소를 맞추고
* 차트의 내부를 채운다

```python
ymin = df['Adj Close'].min()
ymax = df['Adj Close'].max()

plt.ylim(ymin, ymax)
plt.fill_between(df.index, ymin, df['Adj Close'], alpha=0.5)
```
