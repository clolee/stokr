# pandas.read_html
* Read HTML tables into a list of DataFrame objects
* http://goo.gl/TepdQ0 (메뉴얼)

```python
import pandas as pd
from datetime import datetime
```

# 네이버 검색상위
* http://finance.naver.com/sise/lastsearch2.nhn

```python
display(Image('images/naver-top-search.png', width='70%'))
```

```python
url = "http://finance.naver.com/sise/lastsearch2.nhn"

df = pd.read_html(url, header=0)
df = df[1].dropna()
df.set_index('순위', inplace=True)
df.head()
```

```python
now = datetime.now()
df.to_csv(now.strftime("naver-top-search-%Y-%m-%d") + '.csv')
```

#  다음 (DAUM) 조회 급등
* http://finance.daum.net/community/search.daum?page=1
* http://finance.daum.net/community/search.daum?page=2

```python
display(Image('images/daum-top-search.png'))
```

```python
url1 = "http://finance.daum.net/community/search.daum?page=1"
url2 = "http://finance.daum.net/community/search.daum?page=2"

df1 = pd.read_html(url1, header=0)
df2 = pd.read_html(url2, header=0)

df = df1[0].append(df2[0])
df.set_index('순위', inplace=True)
df.head()
```

```python
now = datetime.now()
df.to_csv(now.strftime("daum-top-search-%Y-%m-%d") + '.csv')
```
